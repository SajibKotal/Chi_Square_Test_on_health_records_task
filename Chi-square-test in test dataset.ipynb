# 4. Function for Chi-Square Test

```python
def chi_square_test(data, col1, col2):
    
    clean_data = data.dropna(subset=[col1, col2]) # drop rows where either col1 or col2 has NaN values
    # contingency_table = pd.crosstab(data[col1], data[col2])
    contingency_table = pd.crosstab(clean_data[col1], clean_data[col2])

    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print("\n" + "=" * 60)
    print(f"Chi-Square Test: {col1} vs {col2}")
    print("=" * 60)

    print("\nContingency Table:")
    print(contingency_table)

    print("\nChi-Square Statistic =", round(chi2, 4))
    print("Degrees of Freedom =", dof)
    print("P-value =", p)

    if p < 0.05:
        print("Result: Significant Relationship")
    else:
        print("Result: No Significant Relationship")

    return [col1, col2, chi2, dof, p]
```

# 5. Select Variables (Taget vs other category) for Testing

```python
results = []

tests = [

    ('diabetic', 'profile_hypertensive'),
    ('diabetic', 'has_cardiovascular_disease'),
    ('diabetic', 'had_stroke'),
    ('gender', 'had_stroke'),
    ('gender', 'profile_hypertensive'),
    ('gender', 'diabetic'),
    ('had_stroke', 'has_cardiovascular_disease'),
     # for other features check
    ('RESULT_STAT_BMI', 'had_stroke'),
    ('RESULT_STAT_BMI', 'has_cardiovascular_disease'),
    ('RESULT_STAT_BMI', 'diabetic'),
    ('RESULT_STAT_BMI', 'profile_hypertensive'),
    ('RESULT_STAT_BMI', 'diabetic'),
    ('gender', 'RESULT_STAT_BMI'),
    ('RESULT_STAT_BP', 'had_stroke'),
    ('RESULT_STAT_BP', 'has_cardiovascular_disease'),
    ('RESULT_STAT_BP', 'diabetic'),
    ('RESULT_STAT_BP', 'profile_hypertensive'),
    ('gender', 'RESULT_STAT_BP'),
    ('diabetic', 'RESULT_STAT_SUGAR'),
    ('has_cardiovascular_disease', 'RESULT_STAT_SUGAR'),
    ('had_stroke', 'RESULT_STAT_SUGAR'),
    ('gender', 'RESULT_STAT_SUGAR'),
    ('gender', 'RESULT_STAT_PR'),
    ('diabetic', 'RESULT_STAT_PR'),
    ('had_stroke', 'RESULT_STAT_PR'),
    ('has_cardiovascular_disease', 'RESULT_STAT_PR'),
    ('gender', 'RESULT_STAT_SPO2'),
    ('had_stroke', 'RESULT_STAT_SPO2'),
    ('diabetic', 'RESULT_STAT_SPO2'),
    ('has_cardiovascular_disease', 'RESULT_STAT_SPO2'),
    ('profile_hypertensive', 'RESULT_STAT_MUAC'),
    ('has_cardiovascular_disease', 'RESULT_STAT_MUAC'),
    ('diabetic', 'RESULT_STAT_MUAC'),
    ('had_stroke', 'RESULT_STAT_MUAC'),
    ('gender', 'total_income'),
    ('diabetic', 'total_income'),
    ('had_stroke', 'total_income'),
    ('has_cardiovascular_disease', 'total_income'),
    ('profile_hypertensive', 'total_income'),
    ('Age_Group', 'diabetic'),
    ('Age_Group', 'had_stroke'),
    ('Age_Group', 'has_cardiovascular_disease'),
    ('Age_Group', 'profile_hypertensive'),
    

]
```

```python
for var1, var2 in tests:

    if var1 in df.columns and var2 in df.columns:
        result = chi_square_test(df, var1, var2)
        results.append(result)
```

# 6. Save Results

```python
results_df = pd.DataFrame(
    results,
    columns=[
        'Depentent_Variable_Target',
        'Indepentent_variable',
        'Chi_Square',
        'Degrees_Freedom',
        'P_Value'
    ]
)
```

```python
results_df['Significant'] = results_df['P_Value'] < 0.05

print("\n\nSummary of Results")
print(results_df)
```

```python
results_df.to_excel(
    "chi_square_results.xlsx",
    index=False
)
print("\nResults saved as chi_square_results.xlsx")
```

```python
significant_features = set()

for _, row in results_df.iterrows():

    if row['Significant']:

        significant_features.add(row['Depentent_Variable_Target'])
        significant_features.add(row['Indepentent_variable'])

print("\nStatistically Significant Features:")
print(significant_features)
```
