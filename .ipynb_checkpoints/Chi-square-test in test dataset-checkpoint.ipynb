# 4. Function for Chi-Square Test

```python
def chi_square_test(data, col1, col2):

    contingency_table = pd.crosstab(data[col1], data[col2])

    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print("\n" + "=" * 60)
    print(f"Chi-Square Test: {col1} vs {col2}")
    print("=" * 60)

    print("\nContingency Table:")
    print(contingency_table)

    print("\nChi-Square Statistic =", round(chi2, 4))
    print("Degrees of Freedom =", dof)
    print("P-value =", p)

    if p < 0.05:
        print("Result: Significant Relationship")
    else:
        print("Result: No Significant Relationship")

    return [col1, col2, chi2, dof, p]
```

# 5. Select Variables for Testing

```python
results = []

tests = [

    ('Diabetes', 'Hypertension'),
    ('Diabetes', 'Cardio'),
    ('Diabetes', 'Stroke'),
    ('Gender', 'Stroke'),
    ('Gender', 'Hypertension'),
    ('Gender', 'Diabetes')

]
```

```python
for var1, var2 in tests:

    if var1 in df.columns and var2 in df.columns:
        result = chi_square_test(df, var1, var2)
        results.append(result)
```

# 6. Save Results

```python
results_df = pd.DataFrame(
    results,
    columns=[
        'Variable_1',
        'Variable_2',
        'Chi_Square',
        'Degrees_Freedom',
        'P_Value'
    ]
)
```

```python
results_df['Significant'] = results_df['P_Value'] < 0.05

print("\n\nSummary of Results")
print(results_df)
```

```python
results_df.to_excel(
    "chi_square_results.xlsx",
    index=False
)
print("\nResults saved as chi_square_results.xlsx")
```

```python
significant_features = set()

for _, row in results_df.iterrows():

    if row['Significant']:

        significant_features.add(row['Variable_1'])
        significant_features.add(row['Variable_2'])

print("\nStatistically Significant Features:")
print(significant_features)
```

# 8. Feature Ranking by Significance

```python
ranked = results_df.sort_values(
    by='P_Value',
    ascending=True
)

print("\nFeature Relationships Ranked by Significance:")
print(ranked)
```
